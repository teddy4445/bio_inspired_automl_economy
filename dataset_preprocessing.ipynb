```python
startPath = "data/"
fileName = "2012_Industry_Data_by_Industry_and_StatenoNa"
fileEnd = ".csv"

filePath = startPath + fileName + fileEnd
filePathNoNa = startPath + fileName + "noNa"+ fileEnd
df=pd.read_csv(filePath,thousands = ',')
df.head(20)
```

```python
def clearNa(fileName):
    startPath = "data/"
    fileEnd = ".csv"
    filePath = startPath + fileName + fileEnd
    filePathNoNa = startPath + fileName + "noNa"+ fileEnd
    df=pd.read_csv(filePath)
    df.dropna(axis=1,inplace=True)
    df.to_csv(filePathNoNa,index = False)
```

```python
csvList = ["crypto_BNB_USD","crypto_Cardano_USD","crypto_Dogecoin_USD","crypto_Ethereum_USD","crypto_Litecoin_USD",
           "crypto_Polkadot_USD","crypto_TRON_USD","crypto_XRP_USD"]
for csv in csvList:
    clearNa(csv)
```

```python
def removeK(listToChange):
    #print(type(listToChange.iloc[0]))
    indexOfK = listToChange.str.contains('K')
    indexOfM = listToChange.str.contains('M')
    indexOfM = indexOfM[indexOfM].index
    indexOfK = indexOfK[indexOfK].index
    #print()
    #print(indexOfM)
    #print(listToChange.loc[indexOfK])
    #print(listToChange.loc[indexOfM])
    listToChange[indexOfK] = listToChange[indexOfK].str.replace('K', '')
    listToChange[indexOfM] = listToChange[indexOfM].str.replace('M', '')
    #print(listToChange.loc[indexOfM])
    #print(listToChange)
    #print(listToChange.iloc[592])
    listToChange = pd.to_numeric(listToChange)
    listToChange[indexOfK] = listToChange[indexOfK].mul(1000)
    listToChange[indexOfM] = listToChange[indexOfM].mul(1000000)
    print(listToChange.loc[indexOfM])
    print(listToChange)
    
    
    #print(listToChange)
    return listToChange    
```
