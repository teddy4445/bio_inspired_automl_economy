```python
import pandas as pd
```

```python
df = pd.read_csv('data\Crypto2018-2023.csv')
df
```

```python
uniqueOptions = df.Type.unique()
uniqueOptions
```

```python
grouped = df.groupby(df.Type)
for option in uniqueOptions:
    df1 = grouped.get_group(option)
    df1 = df1.T
    df1 = df1.iloc[1: , :]
    df1.to_csv(f'data\crypto_{option}.csv',index=False)
```

```python
grouped = df.groupby(df.Type)
df1 = grouped.get_group(uniqueOptions[0])
df1 = df1.T
df1
```
